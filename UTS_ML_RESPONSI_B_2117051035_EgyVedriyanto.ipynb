##Library

```python
import numpy as np
import pandas as pd
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
```

##Load Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/credit_dataset.csv")
df
```

##Preprocessing

Menghapus kolom tax_number, tax_cash, dan tax_total

```python
df = df.drop(['tax_number', 'tax_cash', 'tax_total'], axis=1)
df
```

```python
# mengganti nilai NaN pada setiap kolom dengan nilai rata-rata kolom tersebut
df = df.fillna(df.mean())
df
```

```python
df_train = df.drop("customer_id", axis=1)
df_train
```

##Modeling

```python
#Menggunakan K-Means Clustering karena data nya banyak 
#Agglomerative clustering lebih cocok untuk data yang jumlahnya tidak terlalu banyak dan hasil yang lebih kompleks, 
#sementara K-means clustering lebih cocok untuk data yang jumlahnya banyak dan hasil yang lebih sederhana.
n = 2
kmeans_model = KMeans(n_clusters=n, n_init="auto", random_state=100).fit(df_train)
```

```python
#Pelabelan kluster
df_train["kluster"] = kmeans_model.labels_
df_train
```

```python
#Memisahkan fitur dengan target
X = df_train.drop("kluster", axis=1)
y = df_train["kluster"]
```

```python
#Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

#prediksi data test
y_pred = model.predict(X_test)

#menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model:", accuracy*100, "%")
```

##Evaluasi Model

```python
#Evaluasi model menggunakan RandomForest
model = RandomForestClassifier(n_estimators=500, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
#Akurasi baru setelah menggunakan RandomForest
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi :", accuracy*100, "%")
```

1. Silhouette Score
Silhouette score mengukur seberapa baik setiap titik data cocok dengan kelompok yang diberikan berdasarkan jarak antara titik dengan titik pada kelompok yang sama (intra-cluster) dibandingkan dengan jarak dengan titik-titik pada kelompok yang berbeda (inter-cluster). Semakin tinggi nilai silhouette score, semakin baik clustering tersebut.

2. Calinski-Harabasz
Calinski-Harabasz menggunakan rasio antara variansi antara cluster dengan variansi dalam cluster untuk mengevaluasi kualitas clustering. Semakin tinggi nilai Calinski-Harabasz, semakin baik clustering tersebut.

3. Davies-Bouldin
Davies-Bouldin menggunakan rasio antara jarak rata-rata antara setiap pasangan kelompok dengan jarak antara pusat kelompok untuk mengevaluasi kualitas clustering. Semakin rendah nilai Davies-Bouldin, semakin baik clustering tersebut.

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

#Lakukan evaluasi menggunakan silhouette score
df_eval = df_train.copy()

y_predict = kmeans_model.labels_

sil = silhouette_score(df_eval, y_predict)
print(f"Silhouette score : {sil}")
```

```python
#Lakukan evaluasi menggunakan Calinski harabasz
cal = calinski_harabasz_score(df_eval, y_predict)
print(f"Calinski Harabasz score: {cal}")
```

```python
#Lakukan evaluasi menggunakan Davies Bouldin
dav = davies_bouldin_score(df_eval, y_predict)
print(f"Davies Bouldin score: {dav}")
```

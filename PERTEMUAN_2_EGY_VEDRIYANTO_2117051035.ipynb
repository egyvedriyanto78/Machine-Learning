```python
import pandas as pd
import numpy as np

#import apriori from library mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

## Dataset
- Membuat Dataset transaksi sebagai dataframe
- Cek Dataset transaksi

```python
#Membuat data tabular yang terdiri dari data transaksi
df = pd.DataFrame(
    {
        "ID Transaksi": pd.Series([1, 2, 3, 4, 5]),
        "Item": pd.Series([["Susu", "Roti", "Biskuit", "Mentega"], \
                           ["Susu", "Biskuit", "Gula", "Kopi"], \
                           ["Gula", "Kopi", "Susu"], \
                           ["Kopi", "Biskuit", "Susu", "Gula"], \
                           ["Mentega", "Roti", "Kopi"]])
    }
)
```

```python
df.info()
```

```python
df
```

## Encoding data transaksi

```python
df = df.explode('Item')
df
```

```python
basket = (df.groupby(['ID Transaksi', 'Item'])['Item'].count()\
                                              .unstack().reset_index().fillna(0)\
                                              .set_index('ID Transaksi'))
basket.head()
```

```python
#@title
df_encode = df.groupby("Item")["Item"].count().reset_index(name = "Count")
df_encode
```

## Algoritma Apriori

### Apriori Tahap 1

Eliminasi itemset berdasarkan minimum support

```python
#set minimum support 60%
frequent_itemsets = apriori(basket, min_support=0.6, use_colnames=True)

#print(list(frequent_itemsets))
frequent_itemsets
```

### Apriori Tahap 2

Perhitungan Confidence

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules[["antecedents", "consequents", "support", "confidence"]]
```

### Algoritma Apriori (Manual)
- Implementasi algoritma apriori secara manual
- Menggunakan fungsi user defined

#### Data Set

```python
# Inisiasi isi dari Dataset
DATASET = (('Ayam', 'Telur'), \
           ('Ayam', 'Ikan', 'Sapi'), \
           ('Telur', 'Bebek', 'Rusa'), \
           ('Ayam', 'Telur', 'Ikan'), \
           ('Telur', 'Ikan', 'Bebek', 'Kambing'))
```

#### Minimum Support and Confidence

```python
#Inisiasi minimum support
#Minsup 1-5
minsup = int(input("Minimum support\t\t : "))
min_support = minsup

#Inisiasi minimum confidence
#Mincof 0.1 sampai 1.0
mincof = float(input("Minimum confidence\t : "))
min_confidence = mincof
```

#### Iterasi I

```python
#Memanggil library combination
from itertools import combinations as comb

#Membuat variale <frequent> bertipe dictionary
frequent = dict()

#Memasukkan data yang sesuai dengan minimum support ke dalam Frequent
for itemset in DATASET:
    for i in itemset:
        frequent[i] = frequent.get(i,0) + 1
for item in list(frequent):
    if frequent[item]<min_support:
        del frequent[item]
```

```python
#Mulai mengkombinasikan tiap item dan membuang kombinasi yang muncul dibawah 2 atau 40%
items = list(frequent.keys())
support = [frequent]

for i in range(2,len(items)):
    s = dict()
    for combo in comb(items,i):
        for itemset in DATASET:
            if set(combo).issubset(itemset):
                s[combo] = s.get(combo,0) + 1
        if s.get(combo) and s[combo]<min_support:
            del s[combo]
    if not s:
        break
    support.append(s)
support
```

#### Iterasi II

```python
#rules dibuat untuk variable dictionary yang akan menampung confidence dari data set
rules = dict()
for combo in support[-1]:
    for item in combo:
        c = list(combo)
        c.remove(item)
        len_c = len(c)
        c = c[0] if len_c == 1 else tuple(c)

        rule_1 = support[-1][combo]/support[0][item]
        rule1 = "{:.2f}".format(rule_1)

        rule_2 = support[-1][combo]/support[len_c-1][c]
        rule2 = "{:.2f}".format(rule_2)

        if rule_1>=min_confidence: rules[f"{item}->{c}"] = rule1
        if rule_2>=min_confidence: rules[f"{c}->{item}"] = rule2
rules
```
